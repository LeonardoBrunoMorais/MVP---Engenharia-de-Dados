5. Configure as opções de tabela (nome, descrição, etc.) e defina o esquema.

7. Finalize o processo de criação.

### 7.3. Agrupamento dos datasets IMDB

### Descrição
O objetivo deste código é agrupar dados de tabelas que contêm registros das tabelas IMDb Filmes, IMDb Gêneros e IMDb Regiões utilizando a coluna dataId.

### Procedimento
**Mesclando Tabelas**: Para mesclar as informações das tabelas IMDB Filmes, IMDB Gêneros e IMDB Regiões, foi utilizada a função pd.merge() do Pandas para combinar esses DataFrames com base em colunas-chave. te são exibidas para verificação. ###9.6 Modelagem e Inferência

Nessa seção, aplicaremos a etapa de modelagem e inferência, utilizaremos o modelo Regrassão Logisitica. Para isso, iremos utilizar a validação cruzada para treinar e avaliar o modelo, balizada pela métrica de acurácia. Como base nessa técnica, iremos avaliar com o algoritmo com a configuração padrão de hiperparâmetros da biblioteca scikit-Learn.


```
np.random.seed(7) # definindo uma semente global

# Lista que armazenará os modelos
models = []

# Criando o modelo e adicionando-o na lista
models.append(('LR', LogisticRegression(max_iter=200)))

# Definindo os parâmetros do classificador base para o BaggingClassifier
base = DecisionTreeClassifier()
num_trees = 100
max_features = 3

# Criando o modelo para o VotingClassifier
bases = []
model1 = LogisticRegression(max_iter=200)
bases.append(('logistic', model1))
model2 = DecisionTreeClassifier()
bases.append(('cart', model2))
model3 = SVC()
bases.append(('svm', model3))

# Criando os ensembles e adicionando-os na lista de modelos
models.append(('Bagging', BaggingClassifier(base_estimator=base, n_estimators=num_trees)))
models.append(('RF', RandomForestClassifier(n_estimators=num_trees, max_features=max_features)))
models.append(('ET', ExtraTreesClassifier(n_estimators=num_trees, max_features=max_features)))
models.append(('Ada', AdaBoostClassifier(n_estimators=num_trees)))
models.append(('GB', GradientBoostingClassifier(n_estimators=num_trees)))
models.append(('Voting', VotingClassifier(bases)))

# Listas para armazenar os resultados
results = []
names = []

# Avaliação dos modelos
for name, model in models:
    cv_results = cross_val_score(model, X_train, y_train, cv=kfold, scoring=scoring)
    results.append(cv_results)
    names.append(name)
    msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
    print(msg)

# Boxplot de comparação dos modelos
fig = plt.figure(figsize=(10,8))
fig.suptitle('Comparação dos Modelos')
ax = fig.add_subplot(111)
plt.boxplot(results)
ax.set_xticklabels(names)
plt.show()
```

- **Dados Limitados**: Mesmo se você tivesse acesso a dados históricos de vencedores do Oscar e informações de visualização, os conjuntos de dados seriam limitados e não forneceriam informações abrangentes sobre os critérios de seleção da Academia.

Em resumo, prever os vencedores do Oscar com alta precisão usando apenas o número de visualizações e outros atributos é um desafio quase insuperável devido à complexidade e subjetividade envolvidas nas decisões da Academia.
```
# Filtrar apenas os filmes que ganharam o Oscar
filmes_vencedores = df_Oscar_IMDB[df_Oscar_IMDB['winner'] == 1]

# Contar o número de vitórias para cada filme
contagem_vitorias = filmes_vencedores.groupby('film')['winner'].count().reset_index()

# Ordenar os filmes pelo número de vitórias em ordem decrescente
filmes_mais_vencedores = contagem_vitorias.sort_values(by='winner', ascending=False)

# Selecionar os 5 filmes mais vencedores
top_5_filmes_mais_vencedores = filmes_mais_vencedores.head(5)

# Filtrar os dados originais apenas para os 5 filmes mais vencedores
filmes_top_5 = df_Oscar_IMDB[df_Oscar_IMDB['film'].isin(top_5_filmes_mais_vencedores['film'])]

# Calcular a média das classificações (rating) para esses filmes
media_rating = filmes_top_5.groupby('film')['rating'].mean().reset_index()

# Criar um gráfico de barras para comparar a média de rating e número de vitórias
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plotar a média de rating
ax1.bar(media_rating['film'], media_rating['rating'], color='b', alpha=0.7, label='Média de Rating')
ax1.set_xlabel('Filme')
ax1.set_ylabel('Média de Rating', color='b')
ax1.tick_params(axis='y', labelcolor='b')

# Adicionar uma segunda escala no eixo y para o número de vitórias
ax2 = ax1.twinx()
ax2.plot(top_5_filmes_mais_vencedores['film'], top_5_filmes_mais_vencedores['winner'], marker='o', linestyle='-', color='r', label='Número de Vitórias')
ax2.set_ylabel('Número de Vitórias', color='r')
ax2.tick_params(axis='y', labelcolor='r')

# Definir título
plt.title('Comparação entre Média de Rating e Número de Vitórias dos 5 Filmes Mais Vencedores do Oscar')

# Rotacionar os rótulos no eixo x para melhor legibilidade
plt.xticks(rotation=45)

# Adicionar legenda para as linhas
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper left')

# Exibir o gráfico
plt.tight_layout()
plt.show()
```
